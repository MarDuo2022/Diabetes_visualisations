```python
#!/usr/bin/env python

# make sure to install these packages before running:
# pip install pandas


import pandas as pd
from sodapy import Socrata

```

```python
from config import App_token
```

```python
# Unauthenticated client only works with public data sets. Note 'None'
# in place of application token, and no username or password:
client = Socrata("chronicdata.cdc.gov", App_token)

# results filtered by state: Texas and topic: Diabetes
results = client.get("g4ie-h725",locationabbr="TX",topic="Diabetes")

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
```

```python
dia_df = results_df.loc[results_df["topic"]=="Diabetes",:]
dia_df["question"].value_counts()
```
