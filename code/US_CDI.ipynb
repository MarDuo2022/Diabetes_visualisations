```python
#!/usr/bin/env python

# make sure to install these packages before running:
# pip install pandas


import pandas as pd
from sodapy import Socrata

```

```python
from config import App_token
```

```python
# Unauthenticated client only works with public data sets. Note 'None'
# in place of application token, and no username or password:
client = Socrata("chronicdata.cdc.gov", App_token)

# results filtered by state: Texas and topic: Diabetes
results = client.get("g4ie-h725",topic="Diabetes", yearstart="2019", stratificationid1="OVR",limit=100000)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
```

```python
results_df["question"].value_counts()
```

```python
dia_df = results_df.loc[(results_df["question"]=="Mortality due to diabetes reported as any listed cause of death")\
                        |(results_df["question"]=="Mortality with diabetic ketoacidosis reported as any listed cause of death")\
                        |(results_df["question"]=="Prevalence of diagnosed diabetes among adults aged >= 18 years")\
                        |(results_df["question"]=="Adults with diagnosed diabetes aged >= 18 years who have taken a diabetes self-management course"),:]
dia_df
```

```python
mortality_df = dia_df.loc[dia_df["question"]=="Mortality due to diabetes reported as any listed cause of death",:]
mortality_sort = mortality_df.sort_values(["locationabbr","datavaluetype"])
mortality_sort
```

```python
diagnosed_df = dia_df.loc[(dia_df["question"]=="Prevalence of diagnosed diabetes among adults aged >= 18 years")&(dia_df["datavaluetype"]=="Age-adjusted Prevalence"),:]
diagnosed_sort = diagnosed_df.sort_values("locationabbr")
diagnosed_sort
```

```python
dk_df = dia_df.loc[dia_df["question"]=="Mortality with diabetic ketoacidosis reported as any listed cause of death",:]
dk_sort = dk_df.sort_values(["locationabbr","datavaluetype"])
dk_sort
```

```python
# reduce df columns
final_df = diagnosed_sort[["locationabbr","locationdesc","question","datavalueunit","datavaluetype","datavalue"]]
final_df
```

```python
final_df.to_csv('US-CDI_diabetes_prevalence.csv') 
```
