```python
#import depedencies
import pandas as pd
```

```python
#importing the dataFrames in Pandas
states_df = pd.read_csv("../data/statelatlong.csv")
restaurants_df = pd.read_csv("../data/FastFoodRestaurants.csv")
#Rename state names column for simplicity
states_df.rename(columns = {'City':'State'}, inplace = True)
dibeties_df = pd.read_csv("../data/Merged_data/US-CDI_diabetes_prevalence.csv")
```

```python
restaurants_df.head()
```

```python
#Replacing all the symbols to get unique names as one string in names column and moving all vaules to upper case
restaurants_df['name'] = restaurants_df['name'].str.replace("â€™","")
restaurants_df['name'] = restaurants_df['name'].str.replace("'","")
restaurants_df['name'] = restaurants_df['name'].str.replace(" ","")
restaurants_df['name'] = restaurants_df['name'].str.replace(".","")
restaurants_df['name'] = restaurants_df['name'].str.replace("-","")
restaurants_df['name'] = restaurants_df['name'].str.replace("&","AND")
restaurants_df['name'] = restaurants_df['name'].str.upper()
```

```python
#getting total counts of restaurants by state 
restaurants_df2 = restaurants_df['province'].value_counts().rename_axis('State').reset_index(name='counts')
restaurants_df2.sort_values(by='State').set_index('State').head()
```
